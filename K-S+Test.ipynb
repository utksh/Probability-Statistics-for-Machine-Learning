```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#Normal distribution random variables
x = stats.norm.rvs(size = 1000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw =0.5)
plt.show()
```

```python
#Peforming K-S test on the above random variable and comparing it to normal distribution
stats.kstest(x, 'norm')
```

The result gives a statistic value(D(n,m)) of 0.0426 and the pvalue(c(alpha)) of 0.0507. <br>
That means D(n,m) < C(alpha).

Thus we accept Null Hypothesis i.e the data comes from the specified distribution(Normal Distribution)


```python
#Unifrom distribution random variable
y = stats.uniform.rvs(size=1000)
sns.kdeplot(np.array(y), bw = 0.1)
plt.show()
```

```python
stats.kstest(y,'norm')
```

The statistic > pvalue ,
Thus rejecting Null hypothesis(H0) and accepting (H1), which is at least one value does not match the specified distribution.
