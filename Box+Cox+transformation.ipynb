# This notebook explains the Power transformation or the Box cox transformation for a Loggamma distribution.

Box cox transformation is used to tranform a power law distribution to a Normal distribution.

```python
from scipy import stats
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
#A log gamma continuous random variable. (A power law distribution)
x = stats.loggamma.rvs(5, size =500) + 5
#Plotting a Q-Q plot 
prob = stats.probplot (x, dist = stats.norm, plot = ax1)
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution')
plt.show()
```

The plot does not follow the straight line, this shows that log gamma distribution is not Normally distributed 

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
#Box Cox tranformation of the log gamma distribution
xt, _ = stats.boxcox(x)

#Plotting a Q-Q plot
prob = stats.probplot(xt, dist = stats.norm, plot = ax2)

ax2.set_title('Probplot after Box-Cox transformation')
plt.show()
```

After the box cox transformation, the quantiles follow a straight line which shows that the Log gamma distribution no more
follows a power law distribution it is now converted to Normal Distribution.
