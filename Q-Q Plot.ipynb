#    This notebook explains the practical implementation of Q-Q plot

Q-Q plot is a probability plot , which is a graphical method for comparing two probability distribution by plotting there quantiles against each other

```python
import numpy as np
import pylab
import scipy.stats as stats
```

Taking random numbers which hava a normal distribution. <br>
N(0,1) <br>
Mean = 0 <br>
Standard deviation = 1 <br>

```python
std_normal = np.random.normal( loc = 0, scale =1, size = 1000)
```

```python
std_normal
```

```python
#computing percentile from 1 to 100 of the normally distributed data
for i in range(0,101):
    print(i, np.percentile(std_normal,i))
```

```python
#Random data with normal distribution
measurments = np.random.normal(loc = 20, scale = 5, size = 50000)

#Generates a probability plot of sample data against the quantiles of a specified theoretical distribution(here Normal dist.)
stats.probplot(measurments, dist = 'norm', plot = pylab)
pylab.show()
```

We get a straight line between the plot of the quantiles of the two distribution which shows that our data is normally distributed.

```python
#Uniformily distributed data
measurments = np.random.uniform(low = -1, high = 1, size = 10000)

stats.probplot(measurments, dist = 'norm', plot= pylab)
pylab.show()
```

The plot is not a straight line, which is beacuse we have plotted quantiles of Uniformily distributed data vs quantiles of Normally distributed data.

We conclude that, whenever quantiles of two similar distribution are plotted we get a straight line along y = x.
This plot can also be used as test for testing whether a given data is Normally distributed or not.
